# **Feature Engineering Assignment**

**1. What is a parameter?**
A parameter is a value in a model that is learned from data during training, such as weights in a regression or neural network model.

---

**2. What is correlation?**
Correlation is a statistical measure that shows the relationship or association between two variables.

---

**3. What does negative correlation mean?**
Negative correlation means that when one variable increases, the other variable decreases.

---

**4. Define Machine Learning. What are the main components in Machine Learning?**
Machine Learning is a method of teaching computers to learn patterns from data without explicit programming. Its main components are data, model, features, loss function, and optimizer.

---

**5. How does loss value help in determining whether the model is good or not?**
Loss value measures the error between predicted and actual values, and a lower loss indicates a better model.

---

**6. What are continuous and categorical variables?**
Continuous variables are numerical values that can take any value (e.g., height), while categorical variables represent categories or labels (e.g., gender).

---

**7. How do we handle categorical variables in Machine Learning? What are the common techniques?**
Categorical variables are handled by converting them into numerical form using techniques like label encoding and one-hot encoding.

---

**8. What do you mean by training and testing a dataset?**
Training data is used to train the model, while testing data is used to evaluate its performance.

---

**9. What is sklearn.preprocessing?**
It is a module in scikit-learn used for data preprocessing tasks like scaling, normalization, and encoding.

---

**10. What is a Test set?**
A test set is a portion of data used to evaluate the model after training.

---

**11. How do we split data for model fitting (training and testing) in Python?**
We use functions like train_test_split from scikit-learn to divide data into training and testing sets.

---

**12. How do you approach a Machine Learning problem?**
The approach includes data collection, data cleaning, EDA, feature engineering, model selection, training, evaluation, and tuning.

EDA helps understand the data, detect patterns, handle missing values, and identify relationships.

**17. What is causation? Explain difference between correlation and causation with an example.**
Causation means one variable directly affects another, while correlation only shows association. For example, ice cream sales and temperature are correlated, but temperature causes higher sales.

---

An optimizer is an algorithm used to minimize the loss function. Common types include Gradient Descent, SGD, Adam, and RMSprop, which adjust model parameters to improve accuracy.
