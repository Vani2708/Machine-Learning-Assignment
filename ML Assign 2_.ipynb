1. What is Simple Linear Regression (SLR)? Explain its purpose.
Ans:
Simple Linear Regression (SLR)
It is a statistical method that models the relationship between one independent variable (X) and one dependent variable (Y) using a straight line.

Purpose of SLR
To predict the value of the dependent variable and understand the effect of one variable on another.

9. Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.

```python
# Import libraries
from sklearn.linear_model import LinearRegression
import numpy as np

# Example data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable
y = np.array([2, 4, 5, 4, 5])                # Dependent variable

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope (Î²1):", model.coef_[0])
print("Intercept (Î²0):", model.intercept_)
```
